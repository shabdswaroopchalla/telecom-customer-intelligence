# As This 11 customers are having zero tenure in business context these mean latest customers , the right value according to business imputation is zero for total charges

```python
import numpy as np
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

df.loc[df['TotalCharges'].isna(), 'TotalCharges'] = 0
```
